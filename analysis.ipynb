```python
#Data analysis notebook
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
#load the datasets
covid_train=pd.read_csv("Train.csv")
```

```python
#Drop columns with many NaN.This data is redundant.
covid_train=covid_train.drop("Pre-existing symptoms",axis=1)
```

```python
#Drop rows with empty values
covid_train=covid_train.dropna()

```

```python

#Develop correlation between region and covid variant
regions=covid_train['Region']
variants=covid_train['Variant']

#Scatter plot to show distribution of variants in various regions
plt.scatter(regions,variants)
plt.xlabel("Region")
plt.ylabel("Variant")
plt.show()
```

```python
#The plot shows variant 6 from the dataset is only found in europe. Also europe has been exposed
#all 6 variants of covid.
```

```python
#Develop correlation between known symptoms cough and shortness of breath
y1=covid_train['Cough'];
y2=covid_train['Shortness of breath'];
x=covid_train['Variant']

plt.scatter(x,y1,label='Cough',marker="+")
plt.scatter(x,y2,label='Shortness of breath',marker="s")
plt.xlabel("Variants")
plt.ylabel("Shortness of breath and cough severity")
plt.legend(loc="upper left")
plt.show()
```

```python
#Here shortness of breath is a very common symptom found in all 6 variants with high severity
#Cough is a rare symptom however, at high severity for the variants it appears in.
```

```python
#Develop correlation between age and variant
y=covid_train['Age']
x=covid_train['Variant']

plt.bar(x,y)
plt.xlabel("Variants")
plt.ylabel("Age")
plt.show()
```

```python
#The graph shows the first variant affects older age people as well i.e age> 7 years
#The 6th variant mostly affects people from the ages 0 to 20.
```

```python
#Correlation between body aches,nausea and the covid variant
y1=covid_train['body aches']
y2=covid_train['Nausea']
x=covid_train['Variant']


plt.scatter(x,y2,label='Nausea',marker="+")
plt.scatter(x,y1,label='body aches',marker="s")

plt.xlabel("Variants")
plt.ylabel("Severity of symptoms")
plt.title("Correlation between body ache,nausea and covid variant")
plt.legend()
plt.show()
```

```python
#the scatter plot shows the relation between body aches and nausea on variants. Nausea mostly
# falls unser variants 3,5 whereas body aches are widely spread for all 6 variants 
```

```python
#Statistical data analysis
ages=covid_train['Age']
plt.hist(ages)

plt.title("Frequency distribution of Ages")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.show()
```

```python
#Ages between 20-35 are the most commonly affected people by covid 19
standard_deviation=np.std(ages)
mean=np.mean(ages)

print(mean)
```

```python
#The mean is approximately centered around the plot. Therefore, the distribution of ages can be
#a Normal distribution.The data around the mean occurs more frequently than the others.
```

```python
#Finding distribution of the covid variant
variants=covid_train['Variant']
plt.hist(variants)

plt.title("Frequency distribution of covid variants")
plt.xlabel("Variant")
plt.ylabel("Frequency")
plt.show()
```

```python
#The binomial distribution can appromixately fit the above curve. the Variants variable can take
#only 2 outcomes i.e whether the variant is detected(success) or not(failure).
```
